## My Lane as an ML Task

My selected lane is **Refresh / Content Opportunity Scoring**.

This is mainly a **ranking and scoring task** because the main question is: **Which content pages should be reviewed first?**

The system will assign each page a refresh-priority score and rank pages from highest to lowest priority. The output will support SEO editors and content teams in deciding which pages should be reviewed, updated, expanded, merged, or monitored first.


## Target or Proxy

The ideal target would be an **observed future outcome** showing whether a page improved after it was refreshed.

The starter dataset does not contain confirmed post-refresh outcomes, so I should not create a final target directly from a hand-written rule. For this assignment, I will sketch a provisional proxy using historical decline signals only to show what the target column may look like.

The final target should later be based on a measured outcome in a future time window, such as whether impressions, clicks, or sessions recover after review.


```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

target_preview = df[
    [
        "content_id",
        "client_id",
        "impressions_prev_30d",
        "impressions_last_30d",
        "clicks_prev_30d",
        "clicks_last_30d",
        "sessions_prev_30d",
        "sessions_last_30d",
        "trend_direction"
    ]
].copy()

target_preview["provisional_review_flag"] = (
    target_preview["trend_direction"] == "down"
).astype(int)

target_preview.head(10)
```

## Success Metric

The main success metric for this project is **Precision@K**.

SEO teams usually review only a limited number of pages at a time. Therefore, the model should rank the most valuable refresh opportunities near the top of the list.

For example, **Precision@50** measures how many of the top 50 recommended pages are actually good candidates for review. A higher Precision@K means the model is providing more useful recommendations for editors.


## Unit of Analysis

The unit of analysis is **one content page**.

Each row in the dataframe represents one anonymized content page. The columns describe search performance, engagement, freshness, and content characteristics for that page.

The model will generate one refresh-priority score for each content page.


```python
page_df = df[
    [
        "content_id",
        "content_type",
        "search_volume",
        "impressions_90d",
        "clicks_90d",
        "sessions_90d",
        "content_age_days",
        "days_since_last_update",
        "avg_position",
        "trend_direction"
    ]
]

print("Rows:", page_df.shape[0])
print("Columns:", page_df.shape[1])

page_df.head()
```

## Why ML Beats a Fixed Rule

A simple rule, such as selecting every page with a downward trend or every page older than a certain number of days, would ignore many important signals.

Each content page has multiple factors, including search volume, impressions, clicks, sessions, content age, average position, and freshness. These signals interact in different ways and cannot be captured by a single threshold.

Machine learning can learn patterns from multiple features at the same time and rank pages according to their refresh priority. This helps SEO teams focus on the pages with the highest potential impact while still keeping a human reviewer in the decision-making process.
