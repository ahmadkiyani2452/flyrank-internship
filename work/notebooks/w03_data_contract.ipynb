## 2. Fields: feature / label / context / excluded

## Features

- gsc_impressions
- gsc_clicks
- gsc_avg_position
- ga4_pageviews
- ga4_sessions

These features are available before making a decision and can help rank pages for refresh priority.

---

## Label / Proxy

Refresh Priority (Ranking Score)

The goal is to rank content pages that are most likely to benefit from a refresh based on historical search performance.

---

## Context

- report_date
- client_hash_id
- content_hash_id

These fields identify the observation and are used for grouping or filtering, not for prediction.

---

## Excluded

- client_has_gsc
- client_has_ga4
- gsc_data_available
- ga4_data_available

Reason:
These fields are availability or system flags. They describe data collection status rather than content performance and are not used as predictive features.

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

## Data Limits

This dataset contains historical search and analytics measurements only. It cannot explain why a page performed well or poorly, because factors such as content quality, competitor actions, algorithm updates, or manual content changes are not included.

Some clients have incomplete GSC or GA4 history, so data availability differs across clients. Therefore, the results should be used for decision support rather than as proof of cause and effect.
